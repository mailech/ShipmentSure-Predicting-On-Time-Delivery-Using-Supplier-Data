```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Train.csv")
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Reached.on.Time_Y.N',data=df)
plt.title("On-Time vs Delayed Deliveries")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Mode_of_Shipment',hue='Reached.on.Time_Y.N',data=df)
plt.title("Shipment Mode vs Delivery Status")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Product_importance',hue='Reached.on.Time_Y.N',data=df)
plt.title("Product Importance vs Delivery Status")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Warehouse_block',hue='Reached.on.Time_Y.N',data=df)
plt.title("Warehouse Block vs Delivery Status")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Cost_of_the_Product'],bins=20, kde=True)
plt.title("Distribution of Product Cost")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Weight_in_gms'],bins=20, kde=True)
plt.title("Distribution of Product Weight")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Discount_offered'],bins=20, kde=True)
plt.title("Distribution of Discount Offered")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Cost_of_the_Product'])
plt.title("Boxplot of Product Cost")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Weight_in_gms'])
plt.title("Boxplot of Product Weight")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x='Cost_of_the_Product',
    y='Weight_in_gms',
    hue='Reached.on.Time_Y.N',
    data=df,
)
plt.title("Cost vs Weight")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Discount_offered',
    y='Cost_of_the_Product',
    hue='Reached.on.Time_Y.N',
    data=df,
)
plt.title("Discount vs Cost")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(
    df[['Cost_of_the_Product',
        'Weight_in_gms',
        'Discount_offered',
        'Reached.on.Time_Y.N']],
    hue='Reached.on.Time_Y.N'
)
plt.show()
```
