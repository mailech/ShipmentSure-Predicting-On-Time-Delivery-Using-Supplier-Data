skewness and statistical analysis visualization notebook

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Train.csv')
print(df)
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Discount_offered'], bins=20, kde=True)
plt.title("Distribution of the Discount Offered")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Cost_of_the_Product'])
plt.title("Box Plot of Product Cost")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Weight_in_gms'])
plt.title("Box Plot of Shipment Weight")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Reached.on.Time_Y.N', y='Weight_in_gms', data=df)
plt.title("Weight vs On-Time Delivery")
plt.xlabel("Reached on Time (1=Yes, 0=No)")
plt.ylabel("Weight (grams)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Weight_in_gms', y='Discount_offered', hue='Reached.on.Time_Y.N', data=df)
plt.title("Weight vs Discount (Colored by Delivery Status)")
plt.show()
```

```python
mean_value = df['Discount_offered'].mean()
median_value = df['Discount_offered'].median()
mode_value = df['Discount_offered'].mode()[0]

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
skew_value = df['Discount_offered'].skew()
print("Skewness:", skew_value)
```
