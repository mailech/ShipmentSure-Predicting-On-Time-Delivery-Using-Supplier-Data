```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('.././Train.csv')

sns.set(style="whitegrid")

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Reached.on.Time_Y.N', palette='viridis')
plt.title('Distribution of Reached on Time (0 = On Time, 1 = Late)')
plt.savefig('target_distribution.png')

```

```python

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sns.countplot(data=df, x='Warehouse_block', hue='Reached.on.Time_Y.N', ax=axes[0, 0])
axes[0, 0].set_title('Warehouse Block vs Delivery Status')

sns.countplot(data=df, x='Mode_of_Shipment', hue='Reached.on.Time_Y.N', ax=axes[0, 1])
axes[0, 1].set_title('Mode of Shipment vs Delivery Status')

sns.countplot(data=df, x='Product_importance', hue='Reached.on.Time_Y.N', ax=axes[1, 0])
axes[1, 0].set_title('Product Importance vs Delivery Status')

sns.countplot(data=df, x='Gender', hue='Reached.on.Time_Y.N', ax=axes[1, 1])
axes[1, 1].set_title('Gender vs Delivery Status')

plt.tight_layout()
plt.savefig('categorical_vs_target.png')

```

```python

num_features = ['Cost_of_the_Product', 'Discount_offered', 'Weight_in_gms', 'Customer_care_calls']
df[num_features].hist(bins=20, figsize=(15, 10), color='skyblue', edgecolor='black')
plt.suptitle('Distribution of Numerical Features')
plt.savefig('numerical_distributions.png')
```

```python

# 4. Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation = df.drop('ID', axis=1).corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.savefig('correlation_heatmap.png')

```

```python

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.boxplot(data=df, x='Reached.on.Time_Y.N', y='Discount_offered', ax=axes[0])
axes[0].set_title('Discount Offered vs Delivery Status')

sns.boxplot(data=df, x='Reached.on.Time_Y.N', y='Weight_in_gms', ax=axes[1])
axes[1].set_title('Weight in gms vs Delivery Status')

plt.tight_layout()
plt.savefig('numerical_vs_target.png')
```
