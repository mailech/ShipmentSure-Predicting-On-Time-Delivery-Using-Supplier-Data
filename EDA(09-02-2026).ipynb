Importing all the required libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading the csv file

```python
data=pd.read_csv('Train.csv')
```

Visualizing first few rows of the dataset

```python
data.head()
```

Verifying null values

```python
data.isnull()
```

Verifying sum of null values

```python
data.isnull().sum()
```


Visualising all the various plots

```python
#Count Plot

sns.countplot(x=data['Reached.on.Time_Y.N'],color='Blue',data=data)
plt.xlabel('Reached on Time[0-No,1-Yes]',color='Green',fontsize=15)
plt.ylabel('Count for reached and not reached',color='Green',fontsize=15)
plt.title('Count plot for shipments reached and not reached on time',color='Red',fontsize=15,pad=25)
plt.show()
```
