```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Train.csv")
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Reached.on.Time_Y.N', data=df)
plt.title("On-Time Delivery Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='Mode_of_Shipment', hue='Reached.on.Time_Y.N', data=df)
plt.title("Shipment Mode vs Delivery Status")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='Warehouse_block', hue='Reached.on.Time_Y.N', data=df)
plt.title("Warehouse Block Impact")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Reached.on.Time_Y.N', y='Cost_of_the_Product', data=df)
plt.title("Cost vs Delivery Status")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Reached.on.Time_Y.N', y='Discount_offered', data=df)
plt.title("Discount Offered vs Delivery Status")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Reached.on.Time_Y.N', y='Weight_in_gms', data=df)
plt.title("Weight Impact on Delivery")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Reached.on.Time_Y.N', y='Cost_of_the_Product', data=df)
plt.title("Product Cost vs Delivery Outcome")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Customer_care_calls', hue='Reached.on.Time_Y.N', data=df)
plt.title("Customer Calls vs Delivery")
plt.show()
```
