```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.dpi'] = 100
```

```python
df = pd.read_csv('Train.csv')
print(f"Shape: {df.shape}")
df.head()
```

```python
# histograms for numeric features
numeric_cols = ['Customer_care_calls', 'Customer_rating', 'Cost_of_the_Product',
                'Prior_purchases', 'Discount_offered', 'Weight_in_gms']

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle('Distribution of Numeric Features', fontsize=14)

colors = ['#3498db', '#2ecc71', '#e74c3c', '#9b59b6', '#f39c12', '#1abc9c']
for idx, col in enumerate(numeric_cols):
    ax = axes[idx // 3][idx % 3]
    sns.histplot(data=df, x=col, kde=True, ax=ax, color=colors[idx], bins=20)
    ax.set_title(col)

plt.tight_layout()
plt.show()
```

```python
# countplots for categorical features
cat_cols = ['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Categorical Feature Distribution', fontsize=14)

for idx, col in enumerate(cat_cols):
    ax = axes[idx // 2][idx % 2]
    sns.countplot(data=df, x=col, ax=ax, palette='Set2', order=df[col].value_counts().index)
    ax.set_title(col)
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', 
                    (p.get_x() + p.get_width()/2., p.get_height()),
                    ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# correlation heatmap
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=13)
plt.tight_layout()
plt.show()

print("Correlation with target:")
print(corr['Reached.on.Time_Y.N'].drop('Reached.on.Time_Y.N').sort_values(ascending=False))
```

```python
# boxplots by delivery status
features = ['Cost_of_the_Product', 'Discount_offered', 'Weight_in_gms']

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Features vs On-Time Delivery', fontsize=13)

for idx, col in enumerate(features):
    sns.boxplot(data=df, x='Reached.on.Time_Y.N', y=col, ax=axes[idx], palette='Set2')
    axes[idx].set_title(col)
    axes[idx].set_xlabel('0=Late, 1=On-Time')

plt.tight_layout()
plt.show()
```

```python
# target variable distribution
delivery_counts = df['Reached.on.Time_Y.N'].value_counts()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.pie(delivery_counts, labels=['Late (0)', 'On-Time (1)'], 
        autopct='%1.1f%%', colors=['#e74c3c', '#2ecc71'],
        startangle=90, explode=(0.05, 0.05))
ax1.set_title('Delivery Status Split')

sns.countplot(data=df, x='Reached.on.Time_Y.N', palette=['#e74c3c', '#2ecc71'], ax=ax2)
ax2.set_title('Delivery Count')
ax2.set_xlabel('0=Late, 1=On-Time')
for p in ax2.patches:
    ax2.annotate(f'{int(p.get_height())}', 
                 (p.get_x() + p.get_width()/2., p.get_height()),
                 ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()

print(f"Late: {delivery_counts.get(0, 0)} ({delivery_counts.get(0, 0)/len(df)*100:.1f}%)")
print(f"On-Time: {delivery_counts.get(1, 0)} ({delivery_counts.get(1, 0)/len(df)*100:.1f}%)")
```

```python
# delivery status by categorical features
fig, axes = plt.subplots(2, 2, figsize=(13, 9))

for idx, col in enumerate(cat_cols):
    ax = axes[idx // 2][idx % 2]
    sns.countplot(data=df, x=col, hue='Reached.on.Time_Y.N', ax=ax, palette='Set2')
    ax.set_title(f'Delivery by {col}')
    ax.legend(title='Status', labels=['Late', 'On-Time'])

plt.tight_layout()
plt.show()
```
