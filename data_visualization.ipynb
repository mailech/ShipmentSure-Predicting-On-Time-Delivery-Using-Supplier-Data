# Data Visualization - ShipmentSure

Visual exploration of the shipment delivery dataset using Seaborn and Matplotlib.

**Charts covered:**
1. Distribution plots for numeric features
2. Count plots for categorical features
3. Correlation heatmap
4. Box plots by delivery status
5. Pie chart for target variable
6. Pairwise relationships

---

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.dpi'] = 100
print("Libraries loaded!")
```

```python
# Load dataset
df = pd.read_csv('Train.csv')
print(f"Dataset shape: {df.shape}")
df.head()
```

## 1. Distribution Plots for Numeric Features

Histograms show how each numeric feature is distributed across the dataset.

```python
# Numeric columns to plot
numeric_cols = ['Customer_care_calls', 'Customer_rating', 'Cost_of_the_Product',
                'Prior_purchases', 'Discount_offered', 'Weight_in_gms']

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle('Distribution of Numeric Features', fontsize=16, fontweight='bold')

colors = ['#3498db', '#2ecc71', '#e74c3c', '#9b59b6', '#f39c12', '#1abc9c']

for idx, col in enumerate(numeric_cols):
    row = idx // 3
    col_idx = idx % 3
    sns.histplot(data=df, x=col, kde=True, ax=axes[row][col_idx], color=colors[idx], bins=20)
    axes[row][col_idx].set_title(col, fontsize=11)

plt.tight_layout()
plt.show()
```

## 2. Count Plots for Categorical Features

These show the frequency of each category in the categorical columns.

```python
# Categorical columns
cat_cols = ['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Categorical Feature Distribution', fontsize=16, fontweight='bold')

colors_cat = ['#3498db', '#2ecc71', '#e74c3c', '#9b59b6']

for idx, col in enumerate(cat_cols):
    row = idx // 2
    col_idx = idx % 2
    sns.countplot(data=df, x=col, ax=axes[row][col_idx], 
                  palette='Set2', order=df[col].value_counts().index)
    axes[row][col_idx].set_title(f'{col} Distribution', fontsize=11)
    
    # Add count labels on bars
    for p in axes[row][col_idx].patches:
        axes[row][col_idx].annotate(f'{int(p.get_height())}', 
                                     (p.get_x() + p.get_width()/2., p.get_height()),
                                     ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

## 3. Correlation Heatmap

Shows the relationship strength between all numeric features.

```python
# Correlation matrix
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', linewidths=0.5, square=True)
plt.title('Correlation Heatmap of Numeric Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Show correlations with the target variable
print("\nCorrelation with Target (Reached.on.Time_Y.N):")
print("-"*45)
target_corr = corr_matrix['Reached.on.Time_Y.N'].drop('Reached.on.Time_Y.N').sort_values(ascending=False)
print(target_corr)
```

## 4. Box Plots by Delivery Status

Compares key features between on-time (0) and late (1) deliveries.

```python
# Box plots: Key features vs delivery status
features_to_compare = ['Cost_of_the_Product', 'Discount_offered', 'Weight_in_gms']

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Features vs On-Time Delivery', fontsize=14, fontweight='bold')

for idx, col in enumerate(features_to_compare):
    sns.boxplot(data=df, x='Reached.on.Time_Y.N', y=col, 
                ax=axes[idx], palette='Set2')
    axes[idx].set_title(col, fontsize=11)
    axes[idx].set_xlabel('On-Time (0=No, 1=Yes)')

plt.tight_layout()
plt.show()
```

## 5. Pie Chart - Delivery Status Split

```python
# Target variable distribution
delivery_counts = df['Reached.on.Time_Y.N'].value_counts()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Pie chart
colors_pie = ['#e74c3c', '#2ecc71']
ax1.pie(delivery_counts, labels=['Late (0)', 'On-Time (1)'], 
        autopct='%1.1f%%', colors=colors_pie, startangle=90,
        explode=(0.05, 0.05), shadow=True)
ax1.set_title('Delivery Status Distribution', fontsize=12, fontweight='bold')

# Bar chart
sns.countplot(data=df, x='Reached.on.Time_Y.N', palette=colors_pie, ax=ax2)
ax2.set_title('On-Time vs Late Delivery Count', fontsize=12, fontweight='bold')
ax2.set_xlabel('Delivery Status (0=Late, 1=On-Time)')
for p in ax2.patches:
    ax2.annotate(f'{int(p.get_height())}', 
                 (p.get_x() + p.get_width()/2., p.get_height()),
                 ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

print(f"Late deliveries:    {delivery_counts.get(0, 0)} ({delivery_counts.get(0, 0)/len(df)*100:.1f}%)")
print(f"On-time deliveries: {delivery_counts.get(1, 0)} ({delivery_counts.get(1, 0)/len(df)*100:.1f}%)")
```

## 6. Categorical Features vs Target

```python
# Delivery status grouped by categorical features
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
fig.suptitle('Delivery Status by Categorical Features', fontsize=14, fontweight='bold')

for idx, col in enumerate(cat_cols):
    row = idx // 2
    col_idx = idx % 2
    sns.countplot(data=df, x=col, hue='Reached.on.Time_Y.N',
                  ax=axes[row][col_idx], palette='Set2')
    axes[row][col_idx].set_title(f'Delivery by {col}', fontsize=11)
    axes[row][col_idx].legend(title='On-Time', labels=['Late', 'On-Time'])

plt.tight_layout()
plt.show()
```
