```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('./data/Train.csv')

df.info()
```

```python
print(f"Duplicated rows:", df.duplicated().sum())
```

```python
for col in df.select_dtypes(include='object').columns:
    print(f"Unique values in {col}: {df[col].unique()}")
```

```python
print("\n Warehouse Block Counts:")
print(df['Mode_of_Shipment'].value_counts())
print(df['Gender'].value_counts())
print(df['Warehouse_block'].value_counts())
```

```python
df.describe(include='object')
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Product_importance', hue='Reached.on.Time_Y.N',data=df)
plt.title("Product_Importance vs On-Time Delivery")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Warehouse_block', hue='Mode_of_Shipment',data=df)
plt.title("Product_Importance vs On-Time Delivery")
plt.show()
```
