Importing Libraries



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

Importing dataset

```python
data=pd.read_csv('Train.csv')
```

Count Plot

```python


sns.countplot(x=data['Reached.on.Time_Y.N'],color='Blue',data=data)
plt.xlabel('Reached on Time[0-No,1-Yes]',color='Green',fontsize=15)
plt.ylabel('Count for reached and not reached',color='Green',fontsize=15)
plt.title('Count plot for shipments reached and not reached on time',color='Red',fontsize=15,pad=25)
plt.show()
```

Knowing count of how many shipments reached and not reached with respect to Product Importance 

```python
plt.Figure(figsize=(8,8))
sns.countplot(x='Product_importance',hue='Reached.on.Time_Y.N',data=data)
plt.xlabel("Product Importance ['High','Medium','Low']",fontsize=17,color='Green')
plt.ylabel("Count",fontsize=17,color='Green')
plt.title('Count of each product importance of reached and not reached',fontsize=17,color="Red",pad=10)
plt.show()
```

Histogram

```python
sns.histplot(x=data['Discount_offered'],bins=20,kde=True)
```

```python
sns.boxplot(x='Cost_of_the_Product',data=data)
```
