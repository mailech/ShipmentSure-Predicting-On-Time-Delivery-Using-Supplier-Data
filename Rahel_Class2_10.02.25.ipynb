```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Train.csv')
print(df)
```

```python
print(f"Duplicated rows:", df.duplicated().sum())
```

```python
for col in df.select_dtypes(include='object').columns:
    print(f"Unique values in {col}: {df[col].unique()}")
```

```python
df.describe(include='object')
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Product_importance', hue='Reached.on.Time_Y.N',data=df)
plt.title("Product_Importance vs On-Time Delivery")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Warehouse_block', hue='Reached.on.Time_Y.N',data=df)
plt.title("Warehouse Block vs On-Time Delivery")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    x='Warehouse_block',
    hue='Reached.on.Time_Y.N',
    data=df
)

plt.title('Warehouse Block vs On-Time Delivery')
plt.xlabel('Warehouse Block')
plt.ylabel('Number of Shipments')
plt.legend(title='Reached on Time (1 = Yes, 0 = No)')
plt.show()
```
