```python
import pandas as pd

df = pd.read_csv('Train.csv')
df.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Reached.on.Time_Y.N', data=df)
plt.title("On-Time vs Delayed Shipments")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Mode_of_Shipment', hue='Reached.on.Time_Y.N', data=df)
plt.title("Shipment Mode and Delivery Status")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Product_importance', hue='Reached.on.Time_Y.N', data=df)
plt.title("Product Importance and Delivery Status")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Customer_rating', hue='Reached.on.Time_Y.N', data=df)
plt.title("Customer Rating vs Delivery Status")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Mode_of_Shipment', hue='Product_importance', data=df)
plt.title("Shipment Mode vs Product Importance")
plt.show()
```
